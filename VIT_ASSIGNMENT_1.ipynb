```python
#1. Create a pandas dataframe (dataframe ass df) with numpy random values (4 features and 4 observations)
import pandas as pd
import numpy as np
data = np.random.rand(4, 4)
df = pd.DataFrame(data, columns=['Random value 1', 'Random value 2', 'Random value 3', 'Random value 4'])
df
```

```python
#2. Rename the task -1 'df' dataframe column named as 'Random value 1', 'Random value 2','Random value 3','Random value 4'
df.rename(columns={'Random value 1': 'Feature 1', 'Random value 2': 'Feature 2', 'Random value 3': 'Feature 3', 'Random value 4': 'Feature 4'}, inplace=True)
df
```

```python
#3.  Find the descriptive statistics of the 'df' dataframe
statistics = df.describe()
statistics
```

```python
#4. Check for the null values in 'df' and  find the data type of the columns.
null_values = df.isnull().sum()
data_types = df.dtypes
null_values
data_types
```

```python
#5. Display the 'Random value 2' & 'Random value 3' columns with location method and index location method.
random_value_2_loc = df.loc[:, 'Feature 2']
random_value_3_loc = df.loc[:, 'Feature 3']

random_value_2_idx = df.iloc[:, 1]
random_value_3_idx = df.iloc[:, 2]

random_value_2_loc
random_value_3_loc
random_value_2_idx
random_value_3_idx
```
